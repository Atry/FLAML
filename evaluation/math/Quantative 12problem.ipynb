```python
import json
import os

# Create an empty list to store the JSON data
data_list = []

# Specify the directory you want to scan for JSON files
dir_path = 'problems'

# Get a list of all the JSON files in the directory
json_files = [pos_json for pos_json in os.listdir(dir_path) if pos_json.endswith('.json')]

# Iterate over each JSON file
for filename in json_files:
    with open(os.path.join(dir_path, filename)) as json_file:
        # Load the JSON data from the file
        data = json.load(json_file)

        # Append the data to the list
        data_list.append(data)
        print(data["problem"])
        print("Answer: ", data["correct_ans"])
        print("------------------------")

# (Note: When writing code, always print the results.)
```

```python
import os
import json

def load_samples(base_dir, num_samples=10):
    # List of directories to search for .json files
    folders = ["algebra", "number_theory", "counting_and_probability",
               "prealgebra", "intermediate_algebra", "precalculus"]

    samples = {}

    for folder in folders:
        folder_path = os.path.join(base_dir, folder)
        
        # Check if directory exists
        if not os.path.isdir(folder_path):
            print(f"Warning: {folder_path} not found!")
            continue

        # Load each .json file up to num_samples
        for i in range(num_samples):
            file_path = os.path.join(folder_path, f"{i}.json")

            # Check if file exists
            if not os.path.exists(file_path):
                print(f"Warning: {file_path} not found!")
                continue

            with open(file_path, 'r') as file:
                data = json.load(file)

            # Append to the dictionary with a folder-wise key
            if folder not in samples:
                samples[folder] = []
            samples[folder].append(data)

    return samples

# Usage
base_directory = "./300problems/"  # Replace this with your actual path
samples = load_samples(base_directory, num_samples=10)
```

# LangChain


```python
from langchain.chat_models import ChatOpenAI
api_key = open("key_openai.txt", "r").read().strip()
llm = ChatOpenAI(model_name='gpt-4', openai_api_key=api_key)

from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain.tools.python.tool import PythonREPLTool

tools = [PythonREPLTool()]
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)
```

```python
for p in data_list:
    result = agent.run(p['problem'])
    print("********************************************")
    print("********************************************")
```
